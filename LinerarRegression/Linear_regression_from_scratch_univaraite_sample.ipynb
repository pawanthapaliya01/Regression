```python
''' This code is designed from scratch to perform linear regression for an arbitrary number of samples.

    X represents the univariate features.
    y represents the labels.
    m is the weight (slope).
    b is the bias (intercept).

Note: X and y in this code are placeholders and do not represent a real dataset. 
You can replace them with actual data for your specific use case. '''

import numpy as np
import matplotlib.pyplot as plt

def grad_descent(n, m, b, x, y):
    m_grad = 0.0
    for i in range(n):
        m_grad += x[i] * (y[i] - (m * x[i] + b))
    m_grad = (-2 / n) * m_grad
    return m_grad

def bias(n, m, b, x, y):
    b_grad = 0.0
    for i in range(n):
        b_grad += (y[i] - (m * x[i] + b))
    b_grad = (-2 / n) * b_grad
    return b_grad

def loss_function(n, m, b, x, y):
    total_loss = 0.0
    for i in range(n):
        total_loss += (y[i] - (m * x[i] + b)) ** 2
    total_loss = (1 / n) * total_loss  
    return total_loss

# Example input data
X = np.array([1, 2, 3, 4, 5])  
y = np.array([2, 3, 5, 6, 8])  

n = len(X)
m_update = 0.0  
b_update = 0.01  
lr = 0.001  
iterations = 1000  

losses = []  

# Gradient Descent Loop
for i in range(iterations):
    m = grad_descent(n, m_update, b_update, X, y)
    b = bias(n, m_update, b_update, X, y)
    m_update -= lr * m  
    b_update -= lr * b  
    loss = loss_function(n, m_update, b_update, X, y)
    losses.append(loss)  

# For smooth fit
X_smooth = np.linspace(min(X), max(X), 1000)  # 100 points between min and max X
y_smooth = m_update * X_smooth + b_update  # Predicted values

# fit with more points
plt.scatter(X, y, color='red', label="Data Points")  # Original data points
plt.plot(X_smooth, y_smooth, color='blue', label="Fitted Line")  # Smooth fitted line
plt.xlabel("X (Feature)")
plt.ylabel("y (Target)")
plt.title("Linear Regression using Gradient Descent")
plt.legend()
plt.show()
```
